```python
## csv library ---> Most of the time we are going to work with csv files so that why knowledge with csv library functionaltiy matters

import csv
```

```python
with open('example.txt',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Amit',23])
```

```python
with open('example.txt', mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
## Json Library
## When we are going to work with NoSql Database data then the data inside the NoSql Database is going to be their in the Json Format and along with that when we are going to work with json the datasets as well then again these dataset will be consisting of data in the json format itself
```

```python
## When we get the Json Data, we get in a Json string Format so by using json library we can convert our string data into dictionary format for serializing our data
```

```python
import json
```

```python
data={"Name":"Amit","Age":25}
str_json=json.dumps(data)
print(str_json)
print(type(str_json))
```

```python
json_dict=json.loads(str_json)
print(json_dict)
print(type(json_dict))
```

```python
## Regular Expression 
## Regular Expression is used to retrieve info based on a specific pattern or if the pattern matches
```

```python
import re
```

```python
text='Hello My name is Amit and my password is HelloAmit12345'
```

```python
pattern=r'\d+'
match=re.search(pattern,text)
print(match.group())
```
