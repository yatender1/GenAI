# Range()
### Range() function can be used for working as our iterable
### range() takes an input parameter and the first parameter that the range will be taking as the input will be the intial value that your want to print or you want to operate on and then you will be passing the next input after comma as the parameter as the final value and it can take third parameter as well that will the step size that is the difference between the iterating values 
#### range(initial_value,final_value+1,step_size)

```python
# print the numbers from 1 to 50
# 
for i in range(1, 51,1):
    print(i)
```

##### In the range there is never a neccessity for mentioning the step size and similarly if you want to print the numbers or you want to operate on numbers again there is no neccessity for passing the intial value as well if and only if your intial value is 0

```python
# print numbers from 0 to 100
for i in range(101):
    print(i)
```

##### By using range() function we can repreent the numbers in a decrementing manner as well or in a decrementing order as well

```python
# Print the number from 50 to 1 with a gap of 1 value in decrementing order
for i in range(50, 0, -1):
    print(i)
```

```python
# print the number from 50 to 0 with a gap of 5
for i in range(50, -1, -5):
    print(i)
```

# While Loop

#### While loop is being used for iterating over a series of number when a condition is True and we have to initialise a value as well while iterating

```python
## By making the use of while loop we need to print number from 0 to 50
i=0
while i<=50:
    print(i)
    i+=1        
    
```

```python
# We have to print the numbers from 0 to 50 as the even numbers only
for i in range(51):
    if i%2==0:
        print(i)
```

```python
# printing even number using while loop in between 0 to 50
i=0
while i<=50:
    if i%2==0:
        print(i)
    else:
        pass
    i+=1
```

```python
## Infinite loop---> Infinte loop is being used in a way where in the condition is not being met or we are not putting a condition to terminate loop.

i=1
while i>0:
    print(i)
    i+=1
```

## Loop Control Statements

#### Loop control statements are being used for terminating a loop in between when a condition is met or if you want to have some control over your iterating process or while iterating over the series of numbers

##### break ---> break is our first loop control statement which will help to out to terminate the loop when a certain condition is met

```python
## Print the number from 0 to 50 but once number=18 then terminate the loop
for i in range(51):
    if i==18:
        break
    print(i)        
```

#### continue-----> continue is our second loop control statement which helps us in iterating over a series of numbers but helps us with skipping the values in between once we meet a situation

```python
## Print the number from 0 to 50 but only print even number and for odd numbers skip printing
for i in range(51):
    if i%2==0:
        print(i)
    else:
        pass
```

```python
## print the numbers from 0 to 20 but once 17 is is occuring skip that number and continue the loop
for i in range(21):
    if i==17:
        continue    
    print(i)
```

#### Pass ---> pass statement is being for doing nothing

```python
#print the number from 0 to 20 but once 17 is occuring do nothing
for i in range(21):
    if i==17:
        pass    
    print(i)
```

# Nested Loops

#### Nested Loops ---> It is going to have a loop inside a loop

#### f string ---> f string will be helping us out with printing the numbers with the variable value present inside the colons so we can directly represent the variable value inside the colons message itself where variable value will be shown inside the {}

```python
# Nested loops can help us out with iterating over a seres of numbers as with proper combination of iterating number
for i in range(6):
    for j in range(4):
        print(f"i value={i} and j value={j}")
```

```python
## Print the sum of first n natural numbers
n=int(input("Enter the value of n: "))
sum=0
for i in range(1,n+1):
    sum+=i
print(f"The sum of first {n} natural numbers is {sum}")
```

```python
# Check whether the number is prime or not
num=int(input("Enter a number: "))
count=0
for i in range(1,num+1):
    if num%i==0:
        count+=1
if count==2:
    print(f"{num} is a prime number.")
else:   
    print(f"{num} is not a prime number.")

```
