# Conditional Statements

## Contional Statements works like real life scenarios where if a certian condition is meeting then a statement of code or a block of code gets executed

```python
# Eligiblility for driving on meeting a certain condition
## Conditional Statements---> if statements where if a condition is met, a block of code is executed.
age = int(input("Enter your age: "))
if age >= 18:
    print("You are eligible to drive.")
else:
    print("You are not eligible to drive yet.")
```

## Conditional Statements
## if statement
## elif statement
## else statement

```python
## Figure out that who is the tallest among two friends 
height_friend1 = float(input("Enter the height of friend 1 in cm: "))
height_friend2 = float(input("Enter the height of friend 2 in cm: "))   
if height_friend1 > height_friend2:
    print("Friend 1 is taller.")
else:    
    print("Friend 2 is taller.")
```

```python
## figure out who is tallest among three friends    
height1 = float(input("Enter the height of friend 1 in cm: "))
height2 = float(input("Enter the height of friend 2 in cm: "))  
height3 = float(input("Enter the height of friend 3 in cm: "))
if height1 >= height2 and height1 >= height3:
    print("Friend 1 is the tallest.")   
elif height2 >= height1 and height2 >= height3:
    print("Friend 2 is the tallest.")
else:
    print("Friend 3 is the tallest.")
```

## Nested Conditional Statement

## Nested Conditional Statements involve if statements inside another if statements

```python
# Compare the height of three friends and figure out the tallest one using nested if-else statements
height_a = float(input("Enter the height of friend A in cm: "))
height_b = float(input("Enter the height of friend B in cm: ")) 
height_c = float(input("Enter the height of friend C in cm: "))
if height_a >= height_b:
    if height_a >= height_c:
        print("Friend A is the tallest.")
    else:
        print("Friend C is the tallest.")
else:
    if height_b >= height_c:
        print("Friend B is the tallest.")
    else:
        print("Friend C is the tallest.")
    
```

```python
# Whether if the current year is a leap year or not
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 == 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")    
```
