```python
## We are creating a Function to see that how the translation is going to be their from english to hindi language

def translator(English_text):
    """This function is going to be used for converting the english language into hindi language"""
    
    if English_text=="Hello" or English_text=="hello":
        print(f"The translation for {English_text} is Namaste")

    if English_text=="Help" or English_text=="help":
        print(f"The translation for {English_text} is Madad")

    if English_text=="Required" or English_text=="required":
        print(f"The translation for {English_text} is Aavashyak")
```

### Recursive Function

##### By a Recursive Function we can say that a Function is going to call itself again and again

```python
## Factorial of a number
## Factorial of 8  --> 8*7*6*5*4*3*2*1
## Factorial of 0 or Factorial of 1 is 1

def factorial(num):
    if num==0 or num==1:
        print(f'The factorial of {num} is: 1')
    else:
        factorial=1
        for i in range(1,num+1):
            factorial=factorial*i
        print(f"The factorial of num is : {factorial}")    
```

```python
number=int(input("Enter the number:"))
factorial(number)
```

```python
def factorial_recursive(num):
    if num==0 or num==1:
        return 1
    else:
        return num*factorial_recursive(num-1)
```

```python
factorial_recursive(3)
```

```python
## return num* factorial_recursive(num-1)

...
base condition ---> which will stop the loop or which will let the function to stop calling itself in our case, base condition was ---> if i==0 or i==1 --> 1
num*factorial_recursive(num-1)
num==5

5*factorial_recursive(4) --> 4* factorial_recursive(3) ---> 3*factorial_recursive(2) ---> 2*factorial_recursive(1)

...
```

```python
input_recursive=int(input("Enter the value:"))
print("The factorial of the given number is:", factorial_recursive(input_recursive))
```

#### Function with Multiple Parameters

##### Function with Variable length Arguments
###### Positional Arguments--> Positional arguments are being used so that we can pass multiple values inside our function as an argument irrespective of the number of the values we have mentioned as the parameter inside our function

```python
## Function to add multiple numbers irrespective of how many numbers we are passing

def addition_positional(*num):
    add=0
    for i in num:
        add+=i
    print("The addition of the numbers are:",add)
```

```python
addition_positional(2,3,4)
```

##### Keyword Arguments --> Takes the input as the keyword values and irrespective of any number of keyword values you are going to pass it wont matter

```python
def data(**user_data):
    for key,value in user_data.items():
        print(f'{key}:{value}')
```

```python
data(name="Amit",age=25,gender="Male")
```

##### you guys want to pass keyword value pairs as your input along with positional arguments as well

```python
def variable(*positional, **keyword):
    for i in positional:
        print(f'positional values are :', i)
    for key,value in keyword.items():
        print(f'key is :{key} and value is: {value}')
```

```python
variable(1,2,3,4,5,name="Amit",Age=25,Qualification="BTech")
```
