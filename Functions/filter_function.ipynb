## Filter Function

#### Filter Function is going to retrieve only those items which are meeting a certian condition in an iterable

#### The Filter function constructs an Iterator from elements of an iterable for which a condition remains True or for which a Function returns True value

#### Syntax for Filter Function 

list(filter(func, iterable))

```python
## To retreive only those values which are even

def even_function(value):
    if value%2==0:
        return True
```

```python
iterable_list=[1,2,3,4,5,6,7,8,9]
print(list(filter(even_function,iterable_list)))
```

```python
## Solving the similar problem statement with a lambda function

print(list(filter(lambda x:x%2==0,iterable_list)))
```

```python
## I want to retrieve the values which are greater than 5

print(list(filter(lambda x:x>5,iterable_list)))
```

```python
## If I want to retrieve all those values which are greater than 5 as well and even as well
list_items=[1,2,3,4,5,6,78,9,45,6756,23,454,6576,341]
print(list(filter(lambda x:x>5 and x%2==0,list_items)))
```

```python
## Iterating over the Dictionary items

employees= [{'Name':'Amit','Age':23,'Qualification':'Btech'},
{'Name':'Vivek','Age':25,'Qualification':'Bcom'},
{'Name':'Sunil','Age':29,'Qualification':'Electronics'},
{'Name':'Srishti','Age':33,'Qualification':'Btech in Pertroleum'}]
```

```python
def age_greater(employee):
    if employee['Age']>25:
        return True
```

```python
print(list(filter(age_greater,employees)))
```
