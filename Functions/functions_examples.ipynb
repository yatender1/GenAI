```python
## To create a Function that converts temperature from celicus to Fahrenheit to Celcius or vice versa

## if temperature is in celcius, then celcius to fahrenheit conversion is:
# 1 deg celcius=9/5 +32
# 45 deg celcius= 45*9/5 +32

#1 Fahrenheit= 5/9 - 32
#45 deg Fahrenheit= (45-32)*5/9 deg C

def temperature_change(temp_value,unit):
    if unit=='C':
        temperature=temp_value *9/5+32
        print(f"The temperature for {temp_value} deg celcius in Fahrenheit is :", str(temperature)+ "F")

    elif unit=='F':
        temperature=round((temp_value-32)*5/9,2)
        print(f"The temperature for {temp_value} deg Fahrenheit in celcius is :", str(temperature)+ "F")

    else:
        print("Invalid Valid Entered")
```

```python
temperature_change(1,'C')
```

```python
temperature_change(1, 'F')
```

```python
temperature_change(1,'B')
```

```python
## Define a function that would behave as a password strength checker 
### Strong password --->  if your password:
# 1. Length is greater than 8 characters.
# 2. Password must consist of alleast a lowercase character.
# 3. Password must consist of alleast a uppercase character.
# 4. Password must consist of one single digit
# 5. Password must consist of atleast a special character

### Weak password --->
# If a password doesnot match any above condition then it will be considered as a weak password

### islower() --->checks that whether if the current character is a lowercase character or not
### isupper() --->checks that whether if the current character is a uppercase character or not
### is digit() ---> checks that whether if the current character is a digit or not

def password_strength(password):
    if len(password)<8:
        print(f"The entered password {password} is a Weak Password")

    elif  not any(char.islower() for char in password):
        print(f"The entered password {password} is a Weak Password")

    elif  not any(char.isupper() for char in password):
        print(f"The entered password {password} is a Weak Password")

    elif  not any(char.isdigit() for char in password):
        print(f"The entered password {password} is a Weak Password")
    
    elif  not any(char for char in '!,@,#,$,%,^,&,*,(,),_,+'):
        print(f"The entered password {password} is a Weak Password")

    else:
         print(f"The entered password {password} is a Strong Password")
```

```python
password_strength('Hello')
```

```python
password_strength('adsuifda23232#$##ERER')
```

```python
# Palindrome string is going to have characters similar to it from both the sides or even if you will reverse the string still you will be getting the given string itself

# Check that whether if the current string is a palindrome or not
# Reversing a string can be done by --> string[::-1]

def palindrome(string):
    string_value=string.lower().replace(" ","")

    if string_value==string_value[::-1]:
        print("The current string is a palindrome")

    else:
        print("The current string is not a palindrome")
```

```python
palindrome('abA')
```
