## Lambda Functions

### Lambda Functions are the one Liner Functions defined by the Lambda Keyword

#### Syntax for Lambda Function:

#### Lambda arguments: expression

#### Lambda ---> Keyword which simply let the python interpreter know that the current function is our Lambda Function

#### Arguments --> Input values for our function

#### Expression ---> Logic (Function Body)

#### lambda  x:x**2

#### They can have any number of Arguments but only one expression or one Logic
#### They are used for short operations or as Arguments to Higher Order Functions

#### It can be treated as Function without a Name and they dont even require a Return statement

```python
## Function which calculates the square of the numbers
def square(num):
    return num**2
```

```python
print(square(5))
```

```python
## This similar problem can be solved using Lambda Functions as well
lambda_function=lambda x:x**2
```

```python
type(lambda_function)
```

```python
lambda_function(8)
```

```python
product=lambda a,x,y,z:a*x*y*z
print(product(2,9,4,6))
```

```python
## lambda function to calculate the sum of two numbers
sum=lambda x,y:x+y
print(sum(5,6))
```
