# Operators

# Operators helps in applying operation on the given operands

## Types of operators
1. Airthmetic operators 
Addition(+)
Subtraction (-)
Product or Mulitplication (*)
Division (/)
Modulus (%)
Floor (//)
Exponent(a**b)

2. Comparison operators 
Greater than (>)
Less than (<>)
Greater than or equal to (>=)
Less than or equal to (<=)
Equality check (==)
Not equal to (!=)

3. Logical operators 
And  ---> whether if both conditions are true or not
Or ---> whether if any of the operation is true or not
Not ---> negation or to inverse the result of the given operation
