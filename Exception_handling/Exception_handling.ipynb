##### Exception Handling is a way in which if we are getting the Error in programming or if the user entered input caused an error, then instead of throwing an Error we can raise an exception which can let the use know that the input is not correct or the given input can come up as an Error. ##### Exception are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution.

##### except ---> which is used to handle the error in a way such that the program execution should not stop
##### else----> It will be consisting of the code which we want to run if the try block operation is working without throwing any error
##### Finally ----> Finally block gets executed once we are done with everything and even if we are getting an error or not getting an error still the code inside the finally block gets executed

##### if try----> code is throwing an Error---> Except----> code will be running
