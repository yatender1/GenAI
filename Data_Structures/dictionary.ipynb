## Nested Dictionary

## Nested Dictionary will be used for keeping a dictionary inside another dictionary

```python
## I want to store the information for mutliple students like their name, marks, grade

Nested_dictionary={"Student_1":{"Name":"Amit","Marks":84,"Grade":"A"},
                  "Student_2":{"Name":"Vivek","Marks":93,"Grade":"A+"},
                  "Student_3":{"Name":"Vikas","Marks":98,"Grade":"A+"}}
```

```python
Nested_dictionary
```

```python
# Iterate over these Nested Dictionary

for Outer_key,Outer_value in Nested_dictionary.items():
    print(f"Outer Key is :{Outer_key} and Outer value is :{Outer_value}")
    for key_1,value_1 in Outer_value.items():
        print(f"Inner Key is: {key_1} and Inner value is: {value_1}")
```

## Dictionary Comprehension

## Dictionary comprehension helps us with using Looping statement along with Conditional Statement inside our Dictionary

```python
## Syntax for Dictionary Comprehension:
## {expression for item in iterable if condition}

comprehensive={x**2 for x in range(11)}
```

```python
comprehensive
```

```python
comprehensive={x:x**2 for x in range(11)}
```

```python
comprehensive
```

```python
even_dictionary={x:x**2 for x in range(1,11) if x%2==0}
```

```python
even_dictionary
```
