```python
# Nested tuple

Nested_tuple=((1,2,3),(4,17,18),(15,324,124))
```

```python
Nested_tuple[0][:]
```

```python
# Iterating over a Tuple inside a Tuple 

for i in Nested_tuple:
    for j in i:
        print(j, sep=" ")
    print(" ")
```
