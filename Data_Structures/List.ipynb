## List Methods

### 1. append()---> insert the value at the end of the list
### 2. insert()---> insert the value at a specific index location inside our list
### 3. remove()---> remove the value from our list that we are going to pass inside it
### 4. pop()---> remove the value from the list at a specific index location
### 5. index()---> it provides the index of the value that we are going to pass inside it if that value is present inside that list
### 6. sort()---> it is going to help us out with sorting of values in ascending manner
### 7. reverse()---> it can reverse the values present inside our list i.e, value at last index will be at first index now
### 8. count() ---> it counts the number of occurence of a specific value that we are going to pass inside our list and it returns that count of occurence
### 9. clear() --->for removing all fo the values that we are storing inside our list

## List comprehension

### we are going to use looping statement inside the list so that we can store values inside the list in an iterable format and in list comprehension conditional statements are allowed as well



## Syntax for list comprehension is

### [Expression for item in iterable if condition]

### expression --> what we want to print
### for item in iterable --> looping statement
### if condition --> conditional statement

```python
# I need to store values from 1 to 50 inside the list
list_items=[x for x in range(1,51)]
```

```python
list_items
```

```python
list_squared = [items**2 for items in range(1,16)]
```

```python
list_squared
```

```python
# if i am only interested in looking over the square values which are even
even_items=[x**2 for x in range (1,19) if x%2==0]
```

```python
even_items
```

```python
## Functions in list comprehension as well
list_itemss=["Amit","Apple","Banana","Car","Truck","Laptop","Technology"]
```

```python
length_list=[len(word) for word in list_itemss]
```

```python
length_list
```

## Nested List Comprehension

### We will having the combination of multiple lists in our single list itself

```python
list_itemss
```

```python
length_list
```

```python
pair_list=[(i,j) for i in list_itemss for j in length_list]
```

```python
pair_list
```

```python
Ist_1=[1,3,4]
Ist_2=[2,6,9]

paired_list=[(i,j) for i in Ist_1 for j in Ist_2]
```
