# OS Library

##### OS Library is able to interact with our Operating System and can let us work with Multiple Directories and Files

#### getcwd() --> It is capable of providing us with the idea about the current working directory
#### getcwd() --> It stands for current working directory

```python
import os
```

```python
current_path=os.getcwd()
print("The current working directory is :", current_path)
```

```python
##mkdir()---> make directory ---> It will be helping us out with creating the directory

new_path="New_Folder"
os.mkdir(new_path)
print(f"The new directory {new_path} is created successfully")
```

```python
## listdir() ---> Listing all of the directories present ---> List Directories
All_Items=os.listdir('.')
print(All_Items)
```

```python
### Joining multiple Paths Together
## os.path.join()   --> It will be joining multiple paths together

dir_name="Folder_Data"
file_name="File_Data"
Full_path=os.path.join(dir_name, file_name)
print(f"The Full path is:{Full_path}")

```

```python
## I want to join Multiple path with the Current Path

current_complete_path=os.path.join(os.getcwd(),dir_name,file_name)
print(f"The complete path is : {current_complete_path}")
```

```python
## os.path.exists()

given_path=r"c:\Users\Yatender\Desktop\Generative AI Course\Python\app.py"
if os.path.exists(given_path):
    print("Yes this path exists")
else:
    print("This path doesnot exists")
```

```python
## os.path.isfile() ---> This checks that whether if the current path is a file or not
## os.path.isdir()  ---> This checks that whether if the current path is a directory or not

path='new_folder'

if os.path.isfile(path):
    print("The current path belongs to a file")
elif os.path.isdir(path):
    print("The current path belongs to a directory")
else:
    print("The current path belongs to a file")
```

```python
## The complete path <---> Absolute Path ----> os.path.abspath()
## os.path.abspath() ---> It provides us with the complete path of the absolutepath of file or directory

file_name='Total_data.txt'  # Relative path
complete_path=os.path.abspath(file_name)
print(f"The complete path or absolute path for the given file is: {complete_path}")
```
