## File Handling
1. Read the content from My Text file in the system

```python
##open() --> opening your file
## First Parameter --> File_name
## Second Parameter --> Mode
## Mode is going to decide the way in which we are working with this file
## Mode --> r --> Reading the content from the File
## Mode --> w --> Write the content into the File
## Mode --> r+ --> Read the content first and then writing the content into the file
## Mode --> w+ --> Writing the content first and then reading from it
## Reading your file ---> read() ---> Reads the content from the File

f=open('example.txt','r')
content=f.read()
print(content)
f.close()
```

## Writing the content into the file
## write()--> it writes the content in the text file when Mode -->w..> Content will be overwritten and when Mode --> a---> content will be added from the end

```python
f=open('example.txt','w')
f.write('Hello we are now the writing the content into our file')
f.close()
```

```python
## Appending the content into our file

f=open('example.txt','a')
f.write('\nWe are learning about Gen AI pre requisite Python')
f.close()
```

## With  --> It will be used for first of all opening the file as an object and there is no requirement for closing the file as well

```python
with open('example.txt','r') as f:
    content_data=f.read()
    print(content_data)
```

```python
with open('example.txt','w') as file:
    file.write("We are exploring about with command in python File Handling")
```

```python
with open('example.txt','a') as files:
    files.write("\nHello we are exploring addend mode as well")
```

```python
## Write the content into our file line by line ---> writelines()

lines=['First line\n','Second Line\n','Third Line\n']
with open('example.txt','w') as data:
    data.writelines(lines)
```

```python
# r,w,a
## Mode --> Write the content into the binary file --->wb
## Mode --> Read the content from the binary file  --->rb

data=b'\x00\x01\x02\x04\x05\x06\x07\x08\x09\x10\x11\x12'
with open('example.txt','wb') as binary_data:
    binary_data.write(data)
```

```python
with open('example.txt','rb') as binary_read:
    binary_content=binary_read.read()
    print(binary_content)
```

```python
### Read the content from example.txt and writing the content into another file

with open('source.txt','r') as source:
    content=source.read()

with open('destination.txt','w') as destination:
    destination.write(content) 
```

```python
## Reading the Total lines, Total words, Total Characters from a file ---> readlines()
def count_lines(file):
    with open(file, 'r') as files:   # ✅ use variable
        lines = files.readlines()
        total_lines = len(lines)
        total_words = sum(len(line.split()) for line in lines)
        total_characters = sum(len(line) for line in lines)
        return total_lines, total_words, total_characters

file_path = 'Total_data.txt'
Total_Lines, Total_Words, Total_Characters = count_lines(file_path)
print(f"Total Lines are : {Total_Lines} and Total Words are : {Total_Words} and Total Characters are : {Total_Characters}")
```

```python
## Write the content into a file and then after writing the content into that file we are going to read the content from that file as well -->seek()
```

##### w+---> First of all W is their so it states that first writing operation will be done and then reading operation will be done

##### seek() ---> It is going to help us out with reading the content after writing from the starting only

```python
with open('Total_data.txt','w+') as write_read_file:
    write_read_file.write("\n We are looking into W+ operation")
    write_read_file.write("\n It means writing and then Reading")
    write_read_file.seek(0)
    content_complete_data=write_read_file.read()
    print(content_complete_data)
```
